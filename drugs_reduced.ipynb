```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#condense benefits review, comments review, and side effects review into one 
drugs = pd.read_csv("drugs_merged.csv")
drugs["benefitsReview"] = drugs["benefitsReview"].str.lower()
drugs["commentsReview"] = drugs["commentsReview"].str.lower()
drugs["sideEffectsReview"] = drugs["sideEffectsReview"].str.lower()
benefits = drugs[drugs["review"].isnull()].benefitsReview.fillna(" ")
comments = drugs[drugs["review"].isnull()].commentsReview.fillna(" ") 
side_effects = drugs[drugs["review"].isnull()].sideEffectsReview.fillna(" ")
reviews = benefits + " " + comments + " " + side_effects 
drugs.loc[drugs["review"].isnull(), "review"] = reviews 
drugs["review"]  = drugs["review"].str.lower()
drugs[["drugName", "over_counter_name", "condition", "effectiveness", "sideEffects", "price", "rating", "review", "usefulCount"]].to_csv("drugs_reduced.csv")
```
